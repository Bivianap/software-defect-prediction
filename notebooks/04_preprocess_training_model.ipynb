```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
from config import NAME_FILE_BUGS, NAME_FILE_COMMITS, NAME_FILE_METRICS_CODE
from scripts.a_data_merger import perform_merge
from scripts.a_preprocessor import bug_checker, normalization_data


# This file makes the merge between commits, metrics and bugs
###############################################################

mode = 'train'

# path
data_folder = os.path.join(os.getcwd(), '..', 'data', 'extractions')
if not os.path.exists(data_folder):
    os.makedirs(data_folder)

# load data from code --> commits
commit_data_path = os.path.join(data_folder, NAME_FILE_COMMITS)
if os.path.exists(commit_data_path):
    commit_data = pd.read_csv(commit_data_path, sep='|')
else:
    raise FileNotFoundError(f"File not found {commit_data_path}")

# load data from issues --> Jira or azure (only real bugs?: not include duplicated, non-reproducible, wrongly declared, cancelled, evolution future, no correction planned)
# this decision depends on strategy of team
bugs_data_path = os.path.join(data_folder, NAME_FILE_BUGS)
if os.path.exists(bugs_data_path):
    bugs_data = pd.read_csv(bugs_data_path, sep=';')
else:
    raise FileNotFoundError(f"File not found {bugs_data_path}")

# load code metrics
code_data_path = os.path.join(data_folder, NAME_FILE_METRICS_CODE)
if os.path.exists(code_data_path):
    metrics_data = pd.read_csv(code_data_path)
else:
    raise FileNotFoundError(f"File not found {code_data_path}")

##--------------------------------------------------

# merge
merged_data = perform_merge(commit_data, bugs_data, metrics_data, mode)

# bug validation
merged_data = bug_checker(merged_data)

# normalization of data
merged_data = normalization_data(merged_data, mode)




```
