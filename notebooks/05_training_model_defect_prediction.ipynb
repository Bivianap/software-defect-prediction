```python
import sys
import os
sys.path.insert(0, os.path.abspath('..'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_curve, auc, confusion_matrix
from lime.lime_tabular import LimeTabularExplainer
from scipy.optimize import differential_evolution
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier
from scripts.a_results import explainability
from config import UMBRAL

mode = 'train'

# paths
data_folder = os.path.join(os.getcwd(), '..', 'data', 'preprocess')

data_folder_info = os.path.join(os.getcwd(), '..', 'data', 'info')

models_folder = os.path.join(os.getcwd(), '..', 'models')

os.makedirs(models_folder, exist_ok=True)
kmeans_path = os.path.join(models_folder, "kmeans.pkl")
bug_predictor_path = os.path.join(models_folder, 'bug_predictor.pkl')

# load data
X_scaled = pd.read_csv(os.path.join(data_folder, 'final_dataset_complete.csv'))


#define y
y = X_scaled['Bug'] 
X_scaled = X_scaled.drop(columns=['Bug'])

print("Shape of X_scaled:", X_scaled.shape)
print("Length of y:", len(y)) 


# ---------------------- data set preparation ------------------------------

# Divide data in training and test (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Divide the set of training in training and validation (80% training, 20% validation)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

print("Size of training set:", X_train.shape)
print("Size of validation set:", X_val.shape)
print("Size testing set:", X_test.shape)

# apply K-means Clustering to the training set
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train)

# Add cluster labels to each subset
X_train['cluster'] = kmeans.labels_
X_val['cluster'] = kmeans.predict(X_val)
X_test['cluster'] = kmeans.predict(X_test)

joblib.dump(kmeans, kmeans_path)

ruta_archivo = os.path.join(models_folder, "x_train_for_lime.csv")
X_train.to_csv(ruta_archivo, index=False)


# -----------------------Training - models selection for the exercise SVM, NB, random forest, gradient boosting and balanced random forest-------------------------

# SVM
# svm_model = SVC()
# svm_model.fit(X_train, y_train)
# svm_preds = svm_model.predict(X_test)
# print("SVM Accuracy:", accuracy_score(y_test, svm_preds))
# print(classification_report(y_test, svm_preds))

# Naive Bayes
# nb_model = GaussianNB()
# nb_model.fit(X_train, y_train)
# nb_preds = nb_model.predict(X_test)
# print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_preds))
# print(classification_report(y_test, nb_preds))

# Random Forest
brf_model = RandomForestClassifier(random_state=42)
brf_model.fit(X_train, y_train)
brf_probs = brf_model.predict_proba(X_test)  
brf_preds = brf_model.predict(X_test)      
print("Random Forest Accuracy:", accuracy_score(y_test, brf_preds))
print(classification_report(y_test, brf_preds))

# CatBoost model
catboost_model = CatBoostClassifier(random_state=42, verbose=0)
catboost_model.fit(X_train, y_train)
catboost_preds = catboost_model.predict(X_test)
print("CatBoost Accuracy:", accuracy_score(y_test, catboost_preds))
print(classification_report(y_test, catboost_preds))

# Gradient Boosting
# gb_model = GradientBoostingClassifier(random_state=42)
# gb_model.fit(X_train, y_train)
# gb_probs = gb_model.predict_proba(X_test)  
# gb_preds = gb_model.predict(X_test)      
# print("Gradient Boosting Accuracy:", accuracy_score(y_test, gb_preds))
# print(classification_report(y_test, gb_preds))

# Balanced Random Forest
# brf_model = BalancedRandomForestClassifier(random_state=42)
# brf_model.fit(X_train, y_train)
# brf_probs = brf_model.predict_proba(X_test)  
# brf_preds = brf_model.predict(X_test)
# print("Balanced Random Forest Accuracy:", accuracy_score(y_test, brf_preds))
# print(classification_report(y_test, brf_preds))

# decision tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
dt_preds = dt_model.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, dt_preds))
print(classification_report(y_test, dt_preds))

# Multi-layer Perceptron (MLP)
# mlp_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)
# mlp_model.fit(X_train, y_train)
# mlp_preds = mlp_model.predict(X_test)
# print("MLP Accuracy:", accuracy_score(y_test, mlp_preds))
# print(classification_report(y_test, mlp_preds))

#  K-Nearest Neighbor (KNN)
# knn_model = KNeighborsClassifier(n_neighbors=5)  # 5 vecinos por defecto
# knn_model.fit(X_train, y_train)
# knn_preds = knn_model.predict(X_test)
# print("KNN Accuracy:", accuracy_score(y_test, knn_preds))
# print(classification_report(y_test, knn_preds))

#-------------------------------Model Evaluation -------------------------------------------


# 1. Confusion matrix
cm = confusion_matrix(y_test, brf_preds)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predictions')
plt.ylabel('Real values')
plt.title('Confusion matrix')
plt.show()

# 2. Curve ROC y AUC (binary classification)
if hasattr(brf_model, "predict_proba"):  #
    y_probs = brf_model.predict_proba(X_test)[:, 1]  
    fpr, tpr, thresholds = roc_curve(y_test, y_probs)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(10, 7))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curve ROC (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Curve ROC')
    plt.legend(loc="lower right")
    plt.show()
else:
    print("The model doesn't support `predict_proba`, therefore is not possible to generate the curve ROC.")

# 3. Importance of features
feature_importances = brf_model.feature_importances_
features = X_train.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# 4. Grapgh of importance of features
plt.figure(figsize=(10, 7))
sns.barplot(x=importance_df['Importance'][:10], y=importance_df['Feature'][:10], palette='viridis')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Importance of Features')
plt.show()

# 4. Distribution of predictions
plt.figure(figsize=(10, 7))
sns.histplot(brf_preds, bins=3, kde=False, color='blue')
plt.xlabel('Predicted Classes')
plt.ylabel('Frequency')
plt.title('Distribution of prediction')
plt.show()

# 5. Comparison of real values vs predicted
plt.figure(figsize=(10, 7))
plt.scatter(range(len(y_test)), y_test, color='green', alpha=0.5, label='Real Values')
plt.scatter(range(len(brf_preds)), brf_preds, color='red', alpha=0.5, label='Predictions - Balanced Random Forest')
plt.xlabel('Index')
plt.ylabel('Classes')
plt.title('Comparison of real values and predictions')
plt.legend()
plt.show()

# 6. Histogram of probabilities
plt.figure(figsize=(10, 7))
sns.histplot(brf_probs[:, 1], kde=False, color='blue', bins=20)
plt.xlabel('Bug Probability')
plt.ylabel('Frequency')
plt.title('Distribution of probabilities')
plt.show()

# ==============================================================
# LIME (Explicability local) with prioritization on probability
# ==============================================================

print("\n=== LIME for Class 1 'With Bug' order by probability descen ===")

# get the probabilities for class 1 predicted by the model
y_probs_class_1 = brf_model.predict_proba(X_test)[:, 1]

# get binary predictions (0 - 1) using umbral
y_pred_bug = (y_probs_class_1 >= UMBRAL).astype(int)

# how many files have been predicted 'with bug'
quantity_with_bug = np.sum(y_pred_bug)

# print the quantity
print(f"Quantity of files with positive prediction for Bug: {quantity_with_bug} from a total of {len(y_test)}")


# Create a DataFrame with the probabilities, index and values of file_hash
probabilities_df = pd.DataFrame({
    'index': X_test.index,
    'file_hash': X_test['file_hash'].values,
    'probability_class_1': y_probs_class_1
})


# Order by probabilities of class 1 descendent
probabilities_df = probabilities_df.sort_values(by='probability_class_1', ascending=False)
probabilities_df_unique = probabilities_df.drop_duplicates(subset=['file_hash'])

#read file with file_names
file_hashes_df = pd.read_csv(os.path.join(data_folder_info, 'file_hashes_backup.csv'))
file_hashes_df = file_hashes_df.drop_duplicates(subset=['file_name_unique'])

 # Create explainer LIME with all features
explainer_lime = LimeTabularExplainer(
    training_data=np.array(X_train),  # all features
    mode="classification",
    feature_names=X_train.columns, 
    class_names=["Free Bug", "With Bug"], 
    discretize_continuous=True
)

origin_instance = X_test

explainability(probabilities_df_unique, file_hashes_df, mode, brf_model, X_train, origin_instance)

# ==============================================================
#  Differential Evolutionary Optimization, expected values 100-155 and 10-30
# ==============================================================


print("\n=== Differential Evolutionary Optimization Random Forest===")

def objective_function(params):
    n_estimators = int(params[0])
    max_depth = int(params[1])

    model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
    model.fit(X_train, y_train.values.ravel())
   
    preds = model.predict(X_test)
    return -accuracy_score(y_test, preds) 

bounds = [(50, 200),  # n_estimators and max_depth
          (10, 30)]     

result = differential_evolution(objective_function, bounds, strategy='best1bin', maxiter=10)
print("Best parameters found (n_estimators, max_depth):", result.x)

best_n_estimators = int(result.x[0])
best_max_depth = int(result.x[1])

#repeat training with optimized parameters
brf_model = RandomForestClassifier(
    n_estimators=best_n_estimators,
    max_depth=best_max_depth,
    random_state=42
)
brf_model.fit(X_train, y_train)
brf_probs = brf_model.predict_proba(X_test)  
brf_preds = brf_model.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, brf_preds))
print(classification_report(y_test, brf_preds))

#-----------------------------------

print("\n=== Differential Evolutionary Optimization Decision tree===")

def objective_function(params):
    max_depth = int(params[0])
    min_samples_split = int(params[1])

    model = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
    model.fit(X_train, y_train.values.ravel())
    preds = model.predict(X_test)
    return -accuracy_score(y_test, preds)  

bounds = [
    (2, 30),   # max_depth
    (2, 20)    # min_samples_split
]

result = differential_evolution(objective_function, bounds, strategy='best1bin', maxiter=10)
print("Best parameters found (max_depth, min_samples_split):", result.x)

best_max_depth = int(result.x[0])
best_min_samples_split = int(result.x[1])


dt_model = DecisionTreeClassifier(
    max_depth=best_max_depth,
    min_samples_split=best_min_samples_split,
    random_state=42
)
dt_model.fit(X_train, y_train)
dt_probs = dt_model.predict_proba(X_test)
dt_preds = dt_model.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, dt_preds))
print(classification_report(y_test, dt_preds))



# ==============================================================
# Reports
# ==============================================================

#summary report

umbral = UMBRAL
y_pred_bug = (y_probs_class_1 >= umbral).astype(int)
quantity_with_bug = np.sum(y_pred_bug)
cantidad_sin_bug = len(y_pred_bug) - quantity_with_bug

# --- Pie chart ---
labels = ['High probability of Bug', 'Low probability of Bug']
sizes = [quantity_with_bug, cantidad_sin_bug]
colors = ['red', 'green']
explode = (0.1, 0)  

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, startangle=140)
plt.title('Percentage of files with high bug probability')
plt.axis('equal') 
plt.show()

# graph of importance Pie chart
threshold = 0.02  # 2%
selected = importance_df[importance_df['Importance'] > threshold]

# group "others"
rest = importance_df[importance_df['Importance'] <= threshold]
rest_importance = rest['Importance'].sum()

# Prepare labels and size
labels = list(selected['Feature'])
sizes = list(selected['Importance'])

if rest_importance > 0:
    labels.append('Others - Low relevance')
    sizes.append(rest_importance)

# Pie Graph 
colors = sns.color_palette('pastel', n_colors=len(labels))

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Importance of features (>2%) in defect prediction')
plt.axis('equal')
plt.show()

modelo_y_columnas = (brf_model, list(X_train.columns))
joblib.dump(modelo_y_columnas, bug_predictor_path)

print("\nFinished!!")
```
